SUPPORT VECTOR CLASSIFIER IMPLEMENTATION

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#lets create synthetic data points 
from sklearn.datasets import make_classification 
X,y= make_classification(n_samples =1000,n_features= 2,n_classes = 2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(
    x=pd.DataFrame(X)[0],
    y=pd.DataFrame(X)[1],
    hue=y
)
```

```python

from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'linear')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC(kernel = 'rbf')
```

```python
rbf.fit(X_train,y_train)
rbf_pred = rbf.predict(X_test)
```

```python
rbf_pred
```

```python
print(classification_report(y_test,rbf_pred))
print(confusion_matrix(y_test,rbf_pred))
```

```python
poly = SVC(kernel = 'poly')
poly.fit(X_train,y_train)
poly_pred = poly.predict(X_test)
print(classification_report(y_test,poly_pred))
print(confusion_matrix(y_test,poly_pred))
```

```python
sig = SVC(kernel = 'sigmoid')
sig.fit(X_train,y_train)
sig_pred = sig.predict(X_test)
print(classification_report(y_test,sig_pred))
print(confusion_matrix(y_test,sig_pred))
```

#hyperparameter tuning with SVC

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 'C' :[0.1,1,10,100,1000],
              'gamma':[1,0.1,0.01,0.0001],
              'kernel':['rbf']}
```

```python
grid = GridSearchCV(SVC(),param_grid =param_grid,cv=5,verbose =3,refit =True)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(X_test)
```

```python
grid_pred
```

```python
print(classification_report(y_test,grid_pred))
print(confusion_matrix(y_test,grid_pred))
```
